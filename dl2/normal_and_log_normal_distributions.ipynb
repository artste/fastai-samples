# Normal and log of normal 

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
h = stats.norm.pdf(x, mu, sigma)
plt.plot(x, h)
plt.show()
```

```python
hLog = np.log(h)
plt.plot(x, hLog)
plt.show()
```

Log of normal is actually a quadratic function.
